# Convergence Study

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import h5py as h5
import mpl_toolkits
```

```python
def get_number_of_elements(n_elem_multiplier):
    n_ailerons_per_wing = 4
    n_elem_junction_main = int(0.5*n_elem_multiplier)
    if n_elem_junction_main < 1:
        n_elem_junction_main = 1
    n_elem_root_main = int(1*n_elem_multiplier)
    n_elem_tip_main = int(0.5*n_elem_multiplier)
    if n_elem_tip_main < 1:
        n_elem_tip_main = 1
    n_elem_per_aileron = int(4*n_elem_multiplier)
    n_elem_main = int(n_elem_junction_main 
                      + n_elem_root_main 
                      + n_ailerons_per_wing * n_elem_per_aileron 
                      + n_elem_tip_main)
    return n_elem_main
```

```python
def plot_frequencies_modes(matrix_frequencies, x_data, x_label):
    num_modes = matrix_frequencies.shape[0]
    for mode in range(num_modes):
        if not np.all(matrix_frequencies[mode, :] == 0.0):
            plt.plot(x_data, matrix_frequencies[mode, :], '-d')
            plt.title("mode %i" %mode)
            plt.grid()
            plt.xlabel(x_label) # Exchange with num_elements
            plt.ylabel('f [Hz]')
            plt.show()
        else:
            print("For mode %i, the frequency is zero in all cases."%mode)
```

## Wing Only

```python
list_spanwise_multiplier = [1, 2, 4, 8, 16]
list_chordwise_nodes = [4, 8, 16, 32]
output_folder = '../01_case_files/output/'
```

### Structural Convergence (m=const)

```python
array_num_elems = np.zeros((len(list_spanwise_multiplier),1))
num_modes = 20
matrix_frequencies = np.zeros((num_modes, len(list_spanwise_multiplier)))
for counter, multiplier in enumerate(list_spanwise_multiplier):
    file_frequency_table = 'convergence_reference_wing_only_lifting_horseshoe_mult_%im8/beam_modal_analysis/eigenvaluetable.txt' % multiplier
#     file_frequencies = 'convergence_reference_wing_only_lifting_horseshoe_mult_%im8/beam_modal_analysis/frequencies.dat' % multiplier
    
    table = pd.read_csv(os.path.join(output_folder, file_frequency_table), skiprows= [0,2, 23], sep = '|')
    matrix_frequencies[:, counter] = table[table.columns[5]] #np.loadtxt(os.path.join(output_folder, file_frequencies))
    array_num_elems[counter, 0] =get_number_of_elements(multiplier)

plot_frequencies_modes(matrix_frequencies, array_num_elems, 'num_elements')
```

###  Aerodynamic Convergence 

```python

matrix_total_forces = np.zeros((len(list_chordwise_nodes), len(list_spanwise_multiplier), 6))
counter = 0
for counter_multi, multiplier in enumerate(list_spanwise_multiplier):
    for counter_m, num_chordwise_nodes in enumerate(list_chordwise_nodes):
        case = 'convergence_reference_wing_only_lifting_horseshoe_mult_%im%i'%(multiplier, num_chordwise_nodes)
        file_data = case + '/savedata/' + case + '.data.h5'
        try:
            hdf = h5.File(os.path.join(output_folder,file_data), 'r')
            matrix_total_forces[counter_m, counter_multi, :] = np.array(hdf['data']['aero']['timestep_info']['00000']['total_steady_inertial_forces'])[:]
        except:
            print("File not found")
```

```python


fig = plt.figure()
ax = plt.axes(projection='3d')

X, Y = np.meshgrid(list_chordwise_nodes, array_num_elems)
Z = np.transpose(matrix_total_forces[:,:,2])

ax.plot_surface(X, Y, Z)
ax.set_xlabel('M')
ax.set_ylabel('N')
ax.set_zlabel('Fz')

# for counter_multi, multiplier in enumerate(list_mulitplier):
#     for counter_m, num_chordwise_nodes in enumerate(list_chordwise_nodes):
#         plt.plot(matrix_total_forces[counter_m, counter_multi, 2])
```

```python

for counter_multi, multiplier in enumerate(list_spanwise_multiplier):
    plt.plot(list_chordwise_nodes, matrix_total_forces[:, counter_multi, 2], '-d', label = "mutliplier " + str(multiplier))

plt.grid()
plt.legend()
plt.xlabel('N chordwise nodes') # Exchange with num_elements
plt.ylabel('F_z')
plt.show()
#     for counter_m, num_chordwise_nodes in enumerate(list_chordwise_nodes):
#          plt.plot(matrix_total_forces[counter_m, counter_multi, 2])
for counter_nodes, chordwise_nodes in enumerate(list_chordwise_nodes):
    #plt.plot(array_num_elems, matrix_total_forces[counter_nodes, :, 2], label = "M =" + str(chordwise_nodes))
    plt.plot(list_spanwise_multiplier, matrix_total_forces[counter_nodes, :, 2], '-d', label = "M =" + str(chordwise_nodes))

plt.grid()
plt.legend()
plt.xlabel('N spanwise elements') # Exchange with num_elements
plt.ylabel('F_z')
# plt.ylim([570, 670])
plt.show()
```

### Wake

```python
# Horseshoe case as reference
case = 'convergence_reference_wing_only_lifting_horseshoe_mult_%im%i'%(8, 16)
file_horseshoe = '%s/savedata/%s.data.h5'%(case,case)
hd_horseshoe = h5.File(os.path.join(output_folder,file_horseshoe), 'r')
total_forces_horseshoe = np.array(hd_horseshoe['data']['aero']['timestep_info']['00000']['total_steady_inertial_forces'])[:]
```

```python
# Wake cases
list_wake_length = [5, 10, 20, 40]
num_wake_length = len(list_wake_length)
total_forces_wake = np.zeros((num_wake_length, 6))
matrix_frequencies = np.zeros((num_modes, num_wake_length))
for counter, wake_length in enumerate(list_wake_length):
    # Read Data
    case = 'wake_convergence_reference_wing_only_lifting_mult_%im%i_wake%i'%(8, 16, wake_length)
    file_wake = '%s/savedata/%s.data.h5'%(case,case)
    file_frequency_wake ='%s/beam_modal_analysis/eigenvaluetable.txt' % case
    table = pd.read_csv(os.path.join(output_folder, file_frequency_wake), skiprows= [0,2, 23], sep = '|')
    hd_wake= h5.File(os.path.join(output_folder,file_wake), 'r')
    # Save Data in matrices
    total_forces_wake[counter, :] = np.array(hd_wake['data']['aero']['timestep_info']['00000']['total_steady_inertial_forces'])[:]
    matrix_frequencies[:, counter] = table[table.columns[5]] 
```

```python
# plot lift
plt.plot(list_wake_length, total_forces_wake[:, 2], 'd-')
plt.axhline(y = total_forces_horseshoe[2] , xmin = 0, xmax = max(list_wake_length), color ='k', label = "horseshoes")
plt.grid()
plt.show()

#plot frequencies
plot_frequencies_modes(matrix_frequencies, np.dot(list_wake_length,16), 'wake length')
```
